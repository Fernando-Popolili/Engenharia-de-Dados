### Explorando os dados e arrumando as diferenças

```python
colunas_nomes_json = list(dados_json[0].keys())
colunas_nomes_json
```

```python
colunas_nomes_csv = list(dados_csv[0].keys())
colunas_nomes_csv
```

```python
key_map = {
    'Nome do Item' : 'Nome do Produto',
    'Classificação do Produto': 'Categoria do Produto',
    'Valor em Reais (R$)':  'Preço do Produto (R$)',
    'Quantidade em Estoque': 'Quantidade em Estoque',
    'Nome da Loja' : 'Filial', 
    'Data da Venda': 'Data da Venda'
}
```

```python
new_dados_csv = []
for old_dict in dados_csv:
    dict_temp = {} #Dicionario temporario
    print(dict_temp)
    for old_key, value in old_dict.items():
        print(old_key) #Nome do Item
        print(value) #valor Item
        print(key_map[old_key]) #Nome do Produto
        #Dicionario não precisa do "append", apenas dicionario[nome da coluna] = valor            
        dict_temp[key_map[old_key]] = value
        print(dict_temp[key_map[old_key]])
        print(dict_temp)
    new_dados_csv.append(dict_temp)
new_dados_csv[0]
```

```python
#list comprehension
novos_dados_csv = [{key_map.get(old_key): value for old_key, value in old_dict.items()} for old_dict in dados_csv]

novos_dados_csv
```

```python
combined_list = []
combined_list.extend(new_dados_csv)
combined_list.extend(dados_json)




len(combined_list)
```

```python
combined_list
```

```python
nomes_colunas = list(combined_list[0].keys())
nomes_colunas
```

```python
path_dados_combinados = '/home/fe/Documentos/Git/Engenharia-de-Dados/Pipeline Dados/Data Processed/dados_combinados.csv'

with open(path_dados_combinados, 'w') as file:
    writer = csv.DictWriter(file, fieldnames=nomes_colunas)
    writer.writeheader()

    for linha in combined_list:
        writer.writerow(linha)

#Erro devido a coluna 'Data da Venda'
```

### Tratamento da coluna 'Data da Venda' + novo formato para unir as duas tabelas

```python
colunas_nomes = list(combined_list[0].keys())
colunas_nomes
```

```python
dados_combinados = [colunas_nomes]

for row in combined_list:
    linha = []
    for coluna in colunas_nomes:
        print(row.get(coluna))
        linha.append(row.get(coluna, "Indisponivel"))
    dados_combinados.append(linha)
dados_combinados
```

```python
path_dados_combinados= '/home/fe/Documentos/Git/Engenharia-de-Dados/Pipeline Dados/Data Processed/dados_combinados.csv'

with open(path_dados_combinados, 'w') as file:
    writer = csv.writer(file)
    writer.writerows(dados_combinados)

dados_combinados
```
