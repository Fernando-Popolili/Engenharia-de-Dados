# Puxando meu usuario e senha do BD pela .ENV

```python
from dotenv import load_dotenv
import os

load_dotenv('/home/fe/Documentos/Git/Engenharia-de-Dados/03_Pipeline Dados - Integração BD/config/.env')  # caminho do seu arquivo

user = os.getenv("user")
password = os.getenv("password")
```

# Conexão com o MongoDB

```python
#Imports
from pymongo.mongo_client import MongoClient
from pymongo.server_api import ServerApi

#String de conexão
uri = f"mongodb+srv://{user}:{password}@cluster-pipeline.ehjptg7.mongodb.net/?appName=Cluster-pipeline"

# Create a new client and connect to the server
client = MongoClient(uri, server_api=ServerApi('1'))

# Send a ping to confirm a successful connection
try:
    client.admin.command('ping')
    print("Pinged your deployment. You successfully connected to MongoDB!")
except Exception as e:
    print(e)
```

# Criando o BD e uma coleção

```python
db = client['DB_PRODUTOS']
colletion = db['PRODUTOS']
```

```python
client.list_database_names()
```

```python
product = {"Produto": "Computador", "Quantidade": 77}

colletion.insert_one(product)
```

# Extraindo dados da API
